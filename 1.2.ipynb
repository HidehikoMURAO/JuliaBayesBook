### 1.2.1 変数と基本演算

Juliaでは，型の宣言を行わずに直接変数を作成して初期値を設定可能．

```julia
x = 3
y = 1.2
z = "Hi!"
```

typeof 関数を使うことによって，それぞれの変数の型を調べることが可能．

```julia
typeof(x)
```

```julia
typeof(y)
```

```julia
typeof(z)
```

四則演算なども他のプログラミング言語と同様．

```julia
x + 1
```

Unicode 入力が可能
Unicode input https://docs.julialang.org/en/v1/manual/unicode-input/

```julia
# ギリシャ文字
μ = 0.0 #(\mu)
σ = 2.0 #(\sigma)
# 下付き
xᵢ = 0 #(x\_i)
# 上付き
σ² = σ^2 #(\sigma\^2)
```

基本演算の例

```julia
x + y # 加算 addition
```

```julia
x - y # 減産 subtraction
```

```julia
x * y # 乗算 multiplication
```

```julia
x / y # 除算 division
```

```julia
x ^y # べき乗 exponentiation
```

```julia
x ÷ y # 除算の商 quotient
```

```julia
x % y # 剰余 remainder
```

```julia
x == y # 等価 equal
```

```julia
x < y # 小なり less than
```

```julia
x > y # 大なり greater than
```

```julia
 x ≤ y # 小なりイコール less than or equal to (\le, \leq)
```

```julia
x ≥ y # 大なりイコール greater than or equal to (\ge, \geq)
```
